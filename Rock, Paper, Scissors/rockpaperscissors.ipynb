```python
import random  # Import the random module to generate the computer's choice

# List of valid options for the game
options = ['rock', 'paper', 'scissors']

# Ask the user for their choice
user_input = input("Enter rock, paper, scissors, or exit to leave: ").lower()

while user_input != 'exit':

    # Validate the user's input
    if user_input not in options:
        print("Invalid input. Please enter rock, paper, or scissors.")
    
    else:
        # If input is valid, generate the computer's choice
        computer_choice = random.choice(options)

        print(f"\nYou chose: {user_input}")
        print(f"Computer chose: {computer_choice}")

        # Check for a tie
        if user_input == computer_choice:
            print("It's a tie!")

        # Check all winning conditions for the user
        elif (user_input == "rock" and computer_choice == "scissors") or \
             (user_input == "paper" and computer_choice == "rock") or \
             (user_input == "scissors" and computer_choice == "paper"):
            print("You win!")

        # If it's not a tie and not a user win, then the user loses
        else:
            print("You lose!")

    # ASK FOR INPUT AGAIN (THIS FIXES THE INFINITE LOOP)
    user_input = input("\nEnter rock, paper, scissors, or exit to leave: ").lower()

print("\nThanks for playing! Goodbye!")
```
