```python

# ============================================
# 1. Import Libraries
# ============================================
import random
import os

# ============================================
# 2. Create BankAccount Class
# ============================================
class BankAccount(object):
    # ----------------------------------------
    # 2.1 Constructor: create a new account
    # - name: username
    # - accountType: "checking" or "saving"
    # - balance: starting balance (default 0)
    # - generates accountNumber
    # - creates a statement file:
    #   "<accountNumber>_<accountType>_<name>.txt"
    # ----------------------------------------
    def __init__(self, name, accountType, balance=0):
        self.name = name
        self.accountType = accountType
        self.balance = balance

        # Generate random account ID
        self.accountNumber = random.randint(10000, 99999)

        # Build filename using required format
        self.filename = f"{self.accountNumber}_{self.accountType}_{self.name}.txt"

        # Create the file and write initial info
        with open(self.filename, "w") as f:
            f.write(f"Account created for {self.name}\n")
            f.write(f"Account Type: {self.accountType}\n")
            f.write(f"Initial Balance: {self.balance}\n\n")

    # ============================================
    # 3. Deposit Function
    # - Adds money to balance
    # - Logs transaction in statement file
    # ============================================
    def deposit(self, amount):
        self.balance += amount

        with open(self.filename, "a") as f:
            f.write(f"Deposited: {amount}\n")
            f.write(f"New Balance: {self.balance}\n\n")

    # ============================================
    # 4. Withdraw Function
    # - Subtracts money from balance
    # - Prevents withdrawing more than balance
    # - Logs transaction in statement file
    # ============================================
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
            with open(self.filename, "a") as f:
                f.write(f"Failed Withdrawal Attempt: {amount} (Insufficient funds)\n")
                f.write(f"Current Balance: {self.balance}\n\n")
            return

        self.balance -= amount

        with open(self.filename, "a") as f:
            f.write(f"Withdrew: {amount}\n")
            f.write(f"New Balance: {self.balance}\n\n")

    # ============================================
    # 5. Get Balance
    # - Returns current balance
    # ============================================
    def getBalance(self):
        return self.balance

    # ============================================
    # 6. Get User Info Functions
    # - getID: returns accountNumber
    # - getName: returns username
    # - getAccountType: returns account type
    # ============================================
    def getID(self):
        return self.accountNumber

    def getName(self):
        return self.name

    def getAccountType(self):
        return self.accountType

    # ============================================
    # 7. Get Transaction History
    # - Reads and returns full statement file
    # ============================================
    def getHistory(self):
        if not os.path.exists(self.filename):
            return "No transaction history found."

        with open(self.filename, "r") as f:
            return f.read()

# ============================================
# 8. Testing the Class
# - Create multiple accounts
# - Perform deposits and withdrawals
# - Print balances and history
# ============================================

# Create accounts
acc1 = BankAccount("Saad", "checking")
acc2 = BankAccount("Maria", "saving", 500)

# Perform transactions on acc1
acc1.deposit(200)
acc1.withdraw(50)

# Perform transactions on acc2
acc2.deposit(300)
acc2.withdraw(1000)  # Should show "Insufficient funds"

# Print balances
print("acc1 Balance:", acc1.getBalance())
print("acc2 Balance:", acc2.getBalance())

# Print transaction history for acc1
print("\nacc1 Transaction History:")
print(acc1.getHistory())
```
