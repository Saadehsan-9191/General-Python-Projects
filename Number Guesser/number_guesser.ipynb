1. Import libraries


```python
import random
```

2. Guesser Class

```python
class Guesser:
    def __init__(self, number_range=(1, 10), max_guesses=4, max_hints=3):
        self.number = random.randint(number_range[0], number_range[1])
        self.range = number_range
        self.guesses_left = max_guesses
        self.hints_left = max_hints

    def get_factors(self):
        return [i for i in range(1, self.number + 1) if self.number % i == 0]

    def get_multiples(self):
        return [self.number * i for i in range(1, 6)]  # 5 multiples

    def get_larger_smaller(self):
        options = []
        if self.number > self.range[0]:
            options.append("smaller")
        if self.number < self.range[1]:
            options.append("larger")
        return random.choice(options)

    def get_parity(self):
        return "even" if self.number % 2 == 0 else "odd"

    def give_hint(self):
        if self.hints_left == 0:
            return "Sorry, you have no hints left."

        self.hints_left -= 1
        hint_type = random.choice(["a", "b", "c"])

        if hint_type == "a":
            factors = self.get_factors()
            multiples = self.get_multiples()

            if len(factors) > 1:
                return f"Hint: One factor is {random.choice(factors)}"
            else:
                return f"Hint: One multiple is {random.choice(multiples)}"

        elif hint_type == "b":
            choice = self.get_larger_smaller()
            if choice == "larger":
                return f"Hint: The number is larger than {random.randint(self.range[0], self.number - 1)}"
            else:
                return f"Hint: The number is smaller than {random.randint(self.number + 1, self.range[1])}"

        else:
            return f"Hint: The number is {self.get_parity()}"
```

3. Chatbot Interaction

```python
game = Guesser()

print("Welcome to Number Guesser!")
print(f"I have chosen a number between {game.range[0]} and {game.range[1]}.")
print(f"You have {game.guesses_left} guesses and {game.hints_left} hints.")

while game.guesses_left > 0:
    user_input = input("\nEnter your guess or type 'hint': ")

    if user_input.lower() == "hint":
        print(game.give_hint())
        print(f"Guesses left: {game.guesses_left}, Hints left: {game.hints_left}")
        continue

    if not user_input.isdigit():
        print("Please enter a valid number.")
        continue

    guess = int(user_input)

    if guess == game.number:
        print("🎉 Correct! You guessed the number!")
        break
    else:
        game.guesses_left -= 1
        print("Wrong guess. Try again.")
        print(f"Guesses left: {game.guesses_left}, Hints left: {game.hints_left}")

if game.guesses_left == 0:
    print(f"\nYou ran out of guesses. The number was {game.number}.")
```
